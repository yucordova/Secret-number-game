# **Importaciones y base de datos**

```python
#Importaciones
import pandas as pd

# Cargar el dataset
file_path = '/content/drive/MyDrive/Colab Notebooks/Copia de dbip-city-lite-2024-09.csv.gz'
base_de_datos = pd.read_csv(file_path)

# Ver las primeras filas del dataset
print(base_de_datos.head())
```

# **Creación de DataFrame**

```python
#CREAR DATAFRAME
df = pd.DataFrame(base_de_datos)
df
```

```python
#Número de columnas y filas
df.shape
```

```python
#tipo de datos
df.dtypes
```

# **Modificación de las columnas**

```python
#Renombrar columnas
column_name_mapping={
    '0.0.0.0':'primera_direccion_ip',
    '0.255.255.255':'ultima_direccion_ip',
    'ZZ':'codigo_continente',
    'ZZ.1':'nombre_continente',
    'Unnamed: 4':'codigo_iso_3166.alpha2',
    'Unnamed: 5':'Pais_pertenencia_UE',
    '0':'latitud',
    '0.1':'longitud'
}

df_renamed=df.rename(columns=column_name_mapping)
df_renamed
```

```python
#Asignación de variables
dfr=df_renamed
```

# **Contando e indentificando NaNs**

```python
#DETECTAR VALORES NA
dfr.isna()
```

```python
#Conteo de valores NaNs por columna
dfr.isna().sum()
```

```python
# Contar NaNs en cada columna
nan_por_columna = dfr.isna().sum()
print(f"NaNs por columna:\n{nan_por_columna}\n")

print("-"*20)

# Contar NaNs en cada fila
nan_por_fila = dfr.isna().sum(axis=1)
print(f"NaNs por fila:\n{nan_por_fila}")
```

# **Limpieza de NaNs**

```python
# Crear una copia del DataFrame para manipulaciones seguras
df_clean = dfr.copy()
df_clean.head()
```

```python
# Eliminar filas donde cualquier valor es NaN
df_clean.dropna(axis=0, how='any')
```

```python
# Eliminar columnas donde cualquier valor es NaN
df_clean.dropna(axis=1, how='all')
```

```python
#Fill de los NaNs
df_final=df_clean.fillna(
    {
        'codigo_continente': 'Desconocido',
        'nombre_continente': 'Desconocido',
        'codigo_iso_3166.alpha2': 'Desconocido',
        'Pais_pertenencia_UE': 'Desconocido'
    }
)
df_final
```

# **Comprobación de NaNs**

```python
#Conteo de NaNs por columna
df_final.isna().sum()
```

```python
# Contar NaNs en cada columna
nan_por_columna = df_final.isna().sum()
print(f"NaNs por columna:\n{nan_por_columna}\n")

print("-"*20)

# Contar NaNs en cada fila
nan_por_fila = df_final.isna().sum(axis=1)
print(f"NaNs por fila:\n{nan_por_fila}")
```
